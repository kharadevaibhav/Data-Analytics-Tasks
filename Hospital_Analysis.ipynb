```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Data analytics/Hospital_data.csv")
df
```

```python
# Inspect the data
print("HEAD")
display(df.head())
print("\n")
print("\n INFO")
display(df.info())
print('\n')
print("DESCRIBE THE DATA")
display(df.describe())
```

```python
# Checking for null values
null_counts = df.isnull().sum()
null_percentage = (null_counts / len(df)) * 100
display(pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentage}))
```

```python
# Sorting the table if needed (example: sorting by Patient ID if exists)
if 'Patient ID' in df.columns:
    df.sort_values(by='Patient ID', inplace=True)
```

```python
# Insights Dictionary
insights = {}
```

```python
# Analyzing patient data
if 'Readmission' in df.columns:
    readmission_counts = df['Readmission'].value_counts()
    sns.barplot(x=readmission_counts.index, y=readmission_counts.values)
    plt.title("Hospital Readmissions")
    plt.show()
```

```python
# Analyzing Patient Types
if 'Admission Type' in df.columns:
    patient_type_counts = df['Admission Type'].value_counts()
    sns.barplot(x=patient_type_counts.index, y=patient_type_counts.values)
    plt.title("Distribution of Patient Types")
    plt.ylabel("Count")
    plt.xlabel("Patient Type")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Analyzing mortality status
if 'Mortality Status' in df.columns:
    mortality_status_counts = df['Mortality Status'].value_counts()
    sns.barplot(x=mortality_status_counts.index, y=mortality_status_counts.values)
    plt.title("Hospital Mortality Status")
    plt.ylabel("Count")
    plt.xlabel("Mortality Status")
    plt.tight_layout()
    plt.show()
```

```python
# Analyzing medicines and diseases
display(df[['Medicines Prescribed', 'Disease']].dropna().value_counts())
```

```python
# Analyzing annual income
if 'Year' in df.columns and 'Hospital Monthly Income ($)' in df.columns:
    annual_income = df.groupby('Year')['Hospital Monthly Income ($)'].sum()
    sns.lineplot(x=annual_income.index, y=annual_income.values)
    plt.title("Annual Income of Hospital")
    plt.xlabel("Year")
    plt.ylabel("Income")
    plt.show()
```

```python
# Analyzing annual income
if 'Month' in df.columns and 'Hospital Monthly Income ($)' in df.columns:
    annual_income = df.groupby('Month')['Hospital Monthly Income ($)'].sum()
    sns.lineplot(x=annual_income.index, y=annual_income.values)
    plt.title("Annual Income of Hospital")
    plt.xlabel("Month")
    plt.ylabel("Income")
    plt.show()
```

```python
if 'Month' in df.columns and 'Hospital Monthly Income ($)' in df.columns:
    specific_year = 2012


    monthwise_income = df.groupby('Month')['Hospital Monthly Income ($)'].sum()
    sns.barplot(x=monthwise_income.index, y=monthwise_income.values, palette='coolwarm')
    plt.title(f"Month-wise Income in {specific_year}")
    plt.xlabel("Month")
    plt.ylabel("Income")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
